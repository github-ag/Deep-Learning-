```python
import numpy as np
import matplotlib.pyplot as plt
import keras 
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
```

# Doing Image Augmentation
# Benefits of Image Augmentation
- Each time a batch is created while training the data, new augmentation is done which helps to make our model more robust to outliers.
- Deep learning algorithms are hungry for data. So if we have very small dataset it will surely lead to overfitting . So Data Augmentation helps us to create large datasets.

```python
train_image_gen = ImageDataGenerator(
    rescale = 1/255.0, # To get all the pixels in the range 0 to 1.
    rotation_range=40, # rotating image by 40 degrees.
    width_shift_range = 0.2, # shifting all the pixels to 20% left
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.3,
    horizontal_flip=True,
)
# Image Augmentation is done only for training data and not for testing data.
test_image_gen = ImageDataGenerator(rescale=1/255.0)

    
    
```

```python
imgPath = 'Images/My Image/Ishu.jpg'
img = image.load_img(imgPath,target_size=(100,100))
```

### We can load image like this and then can convert it into the array.

```python
img = image.img_to_array(img)
```
